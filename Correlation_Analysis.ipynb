üß† What is Correlation?

Correlation tells you how strongly two numeric values move together.

üìà If both go up together ‚Üí Positive correlation (close to +1)

üìâ If one goes up and the other goes down ‚Üí Negative correlation (close to -1)

‚ùå No relationship ‚Üí Correlation ‚âà 0

üîç Where Do We Use This in Data Science?

‚úÖ To see if features are related

‚úÖ To drop redundant columns

‚úÖ To detect possible hidden patterns

‚úÖ To help in feature selection for ML models


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Recommended Dataset: Iris Dataset
The Iris dataset is perfect because:

It contains only numeric features: sepal length, sepal width, petal length, petal width.

It‚Äôs small and clean.

It‚Äôs used in ML and EDA demos worldwide.

Let‚Äôs load it using seaborn:

```python
# Load Iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

Calculate the Correlation Matrix

```python
corr_matrix = iris.corr(numeric_only = True)
corr_matrix
```

 Visualize the Correlation Matrix with a Heatmap

```python
plt.figure(figsize=(8,6)) ## Set figure size
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidth=0.5)

plt.title("Correlation Heatmap - Iris Dataset")
plt.show()

#plt.figure(figsize=(8, 6)) ‚Üí Makes the heatmap larger and readable.
#sns.heatmap(...) ‚Üí Draws the heatmap.
#annot=True ‚Üí Shows the correlation numbers inside each cell.
#cmap='coolwarm' ‚Üí Uses a blue-to-red color palette (blue = negative, red = positive).
#linewidths=0.5 ‚Üí Adds space between boxes for clarity.
#plt.title() ‚Üí Adds a title to your chart.
#plt.show() ‚Üí Displays the plot.
```
