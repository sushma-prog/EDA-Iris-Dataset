Explore how different variables relate to each other visually using pair plots and other slightly advanced Seaborn tools.

We'll answer:

Do features cluster by species?

Do some features separate categories clearly?

Can we guess which features are more useful for classification?

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Load Iris Dataset

```python
iris = sns.load_dataset("iris")
iris.head()
```

Create a Pair Plot

```python
sns.pairplot(iris, hue="species")
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()

#sns.pairplot() → To see relationships between all numeric features at once.
#hue= → To color by a category (e.g., species).
#hue='species' adds different colors for each flower species.
#This gives scatter plots for each feature pair and histograms on the diagonal.
#By default, the title is placed just above the plot area (at y=1.0). Setting y=1.02 moves the title slightly higher (2% above the normal top), making it clear and well-positioned.
```

---
**Scatter Plot**
  
🔍 1. Setosa stands out clearly:

In almost every scatter plot, the blue dots (setosa) are separated from the others.

Especially in petal_length and petal_width, setosa is completely isolated.

👉 This means Setosa is easy to classify.

🔍 2. Versicolor vs. Virginica:

The orange (versicolor) and green (virginica) dots overlap more, especially in:

sepal_length

sepal_width

But they separate well in:

petal_length

petal_width

👉 So petal-based features are better for classification.

---
**Diagonal Histograms**

What to Look For in These?
                                    
1. Are the curves separated or overlapping?
                                    
✅ Separated = good feature for classification

❌ Overlapping = less useful feature

🔍 Example: Let’s Decode petal_width from diagonal
                                    
You saw three distinct curves, each colored by species.

The blue curve (setosa) is far left.

The orange (versicolor) is in the middle.

The green (virginica) is on the right.

➡️ That means:

Each species has a different petal width range.

So if I know the petal width, I can almost guess the species.

This makes petal_width a powerful feature for classification.

---

❌ Counter Example: sepal_width
                            
In the diagonal histogram for sepal_width, the curves overlap a lot.

You can't easily tell species apart using this.

Diagonal histograms show how values of a feature are spread for each class. If the color curves are separate, it's a great feature. If they overlap, it's not very helpful.

---

1. Peak Height = Frequency (How Common)
2. 
📌 A tall peak means:

Many values are concentrated around that number.

That value is very common.

Example:
If the petal_length for setosa has a tall peak around 1.5, that means most of the setosa flowers have petal lengths close to 1.5.

📌 A short (flat) curve means:

Values are spread out.

No single value is very common.

The feature has high variation.

➡️ 2. Width of the Curve = Spread or Variability

📌 A narrow curve means:

All values are very close together (less variation).

The feature is consistent across data points.

📌 A wider curve means:

Values are more spread out (more variation).

The feature varies a lot in that group.

🔁 3. Multiple Peaks = Possibly Subgroups

If you see two or more bumps (peaks) in one color:

That group might have internal subgroups or clusters.

Something interesting is going on — maybe different species mixed or hidden patterns.

---

✅ Final Intuition Table:

Curve Shape	Meaning

Tall, thin peak	Most values are similar — low variation

Short, wide curve	Values are spread out — high variation

Multiple peaks	May indicate subgroups/clusters or noise

No clear peak	Data is scattered/random, not useful for classification
